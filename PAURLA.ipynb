# **Gráficos**

Colocamos los valores de nivel de educación de los padres en ese orden para que vayan de menor grado a mayor grado

```python
grafico1 = px.bar(dt,'parental level of education',
                  category_orders={'parental level of education': ['some high school','high school','some college',"associate's degree","bachelor's degree","master's degree"]}
                  )
grafico1.show()
```

```python

grafico2 = px.pie(dt,'test preparation course')
grafico2.show()
```

```python
df_melted = dt.melt(id_vars='test preparation course',
                    value_vars=['math score', 'reading score', 'writing score'],
                    var_name='Asignatura',
                    value_name='Nota',)

px.box(df_melted,
        x='Asignatura',               
        y='Nota',                      
        color='test preparation course',
        title='Distribución de Notas Según Curso de Preparación',
        )
```

```python
df_melted = dt.melt(id_vars=['test preparation course', 'parental level of education'],
                    value_vars=['math score', 'reading score', 'writing score'],
                    var_name='Asignatura',
                    value_name='Nota',)

px.box(df_melted,
        x='Asignatura',               
        y='Nota',                      
        color='test preparation course',
        facet_col='parental level of education',
        facet_col_spacing=0.01,
        title='Distribución de Notas Según Curso de Preparación y Educación de los Padres',
        category_orders={'parental level of education': ['some high school','high school','some college',"associate's degree","bachelor's degree","master's degree"]}
        )       
```

```python
df_melted = dt.melt(id_vars=['test preparation course', 'parental level of education','gender','race/ethnicity'],
                    value_vars=['math score', 'reading score', 'writing score'],
                    var_name='Asignatura',
                    value_name='Nota',)

px.box(df_melted,
        x='Asignatura',               
        y='Nota',                      
        color='test preparation course',
        facet_col='race/ethnicity',
        title='Distribución de Notas Según Grupo étnico',
        category_orders={'race/ethnicity': ['group A','group B','group C',"group D","group E"]}
        )       
```
