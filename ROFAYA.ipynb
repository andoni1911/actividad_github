## 5. Análisis descriptivo
### a)Medidas de Tendencia Central:
- Media: valor central que representa el reparto equitativo de un conjunto de datos

```python
Media = dat.mean(numeric_only=True)
print(Media)
```

- Moda: valor o dato que aparece con mayor frecuencia dentro de un conjunto de datos, siendo la medida de tendencia central que indica el elemento más común o repetido.

```python
Moda = dat.mode().iloc[0]
print(Moda)
```

- Mediana: valor que divide un conjunto de datos en dos partes iguales cuando los datos están ordenados de menor a mayor

```python
Mediana = dat.median(numeric_only=True)
print(Mediana)
```

### b)Medidas de Dispersión:
- Rango: diferencia entre el valor máximo y el valor mínimo de un conjunto de datos, y se calcula restando el valor más pequeño al más grande.

```python
rango = dat.max(numeric_only=True) - dat.min(numeric_only=True)
print(rango)

```

- Varianza: medida clave de dispersión que indica qué tan separados o agrupados están los datos de un conjunto con respecto a su media (promedio).

```python
Varianza = dat.var(numeric_only=True).round(3)
print(Varianza)
```

- Desviación estandar: medida estadística de la dispersión o variabilidad de un conjunto de datos, indicando qué tan esparcidos están los valores con respecto a su media (promedio)

```python
Desviacion_estandar = dat.std(numeric_only=True).round(3)
print(Desviacion_estandar)
```

- Rango intercuartílico (IQR): medida de dispersión estadística que indica la variabilidad del 50% central de un conjunto de datos

```python
Q1 = dat.quantile(0.25, numeric_only=True)
Q3 = dat.quantile(0.75, numeric_only=True)
IQR = Q3 - Q1
print(IQR)
```

### c) Medidas de posición:
- Percentiles: medidas de posición que dividen un conjunto de datos ordenados en 100 partes iguales, y cada percentil representa un punto por debajo del cual se encuentra un porcentaje específico de los datos

```python
percentiles = dat.quantile([0.10, 0.25, 0.50, 0.75, 0.90], numeric_only=True)
print (percentiles)
```

### d)Representación gráfica
Instalamos librerias de Seaborn y Matplotlib con "pip install seaborn matplotlib" e importamos las librerias.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

1. HISTOGRAMA (calories burn): representación gráfica de la distribución de frecuencias de datos numéricos continuos.

```python
plt.figure(figsize=(10, 6)) 
sns.histplot(dat['Calories Burn'], kde=True)
plt.title('Distribución de Calorías Quemadas', fontsize=15)
plt.xlabel('Calorías')
plt.ylabel('Frecuencia (Cantidad de personas)')
plt.show()
```

2. GRAFICO DE BARRAS (exercise): epresentación visual que usa barras rectangulares (verticales u horizontales) para comparar frecuencias o valores de variables cualitativas o discretas

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Exercise', data=dat)
plt.title('Frecuencia de Tipos de Ejercicio', fontsize=15)
plt.xlabel('Tipo de Ejercicio')
plt.ylabel('Cantidad de Registros')
plt.xticks(rotation=45)
plt.show()
```

3. BOXPLOT (heart rate/gender): herramienta estadística que visualiza la distribución de un conjunto de datos numéricos, mostrando la mediana, los cuartiles (Q1 y Q3) que definen la caja (el 50% central de los datos), y los "bigotes" que se extenderían hasta los valores extremos (mínimo y máximo), identificando outliers (valores atípicos) con puntos individuales

```python
plt.figure(figsize=(8, 6))

sns.boxplot(x='Gender', y='Heart Rate', data= dat)

plt.title('Análisis de Frecuencia Cardíaca por Género (Detección de Outliers)', fontsize=14)
plt.show()
```
